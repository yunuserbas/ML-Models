**In scikit-learn, import the Iris dataset, it is a great classic of data. It contains 3 species of flowers (irises) according to their length and width of petal and sepal.**

```python
import pandas as pd
import numpy as np
import seaborn as sns

iris = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv")
iris.head()
```

```python
iris['species'].value_counts()
```

```python
iris.select_dtypes(include="number")
```

**Train your model to find 3 clusters, from the 4 numeric columns.**

```python
from sklearn.cluster import KMeans

# Create an instance of the KMeans class
X = iris.select_dtypes(include="number")
modelKM = KMeans(n_clusters=3)

# Fit the model to the data
modelKM.fit(X)

# Get the cluster labels
clusters = modelKM.labels_
clusters
```

```python
modelKM.cluster_centers_
```

**Produce a visualization comparing the basic dataset with the classes generated by the K-means using this image as inspiration. You can display 2 features of your choice among the 4.**
*   On the graph on the left, you have to display the points with different colors for each cluster you belong to, as well as the center of each cluster prominently.
*   On the right graph, you must display the points with different colors corresponding to their iris species.


```python
import matplotlib.pyplot as plt

# Get the species labels for the original dataset
species = iris['species']

# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Scatter plot of sepal length and width with different colors for each cluster
ax1.scatter(iris['sepal_length'], iris['sepal_width'], c=clusters, cmap='rainbow')
# Display the center of each cluster prominently
ax1.scatter(modelKM.cluster_centers_[:, 0], modelKM.cluster_centers_[:, 1], c='black', marker='x')
ax1.set_xlabel('Sepal Length')
ax1.set_ylabel('Sepal Width')
ax1.set_title('3 Clusters')


# Scatter plot of sepal length and width with different colors for each species
ax2 = sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=iris)

ax2.set(xlabel='Sepal Length', ylabel='Sepal Width')
ax2.set_title('3 original species')

plt.show()

```

**Evaluate your model using the Elbow method, between 2 and 10 clusters, which cluster number seems optimal?**

```python
wcss = []
for i in range(2, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(2, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

**Evaluate your model using the Silhouette method, between 2 and 10 clusters, which cluster number seems optimal? And you, visually, if you didn't know there were 3 species, how many would you have found?**

```python
from sklearn.metrics import silhouette_score

silhouette_scores = []
for n_clusters in range(2, 11):
    kmeans = KMeans(n_clusters=n_clusters, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(X)
    silhouette_scores.append(silhouette_score(X, kmeans.labels_))

plt.plot(range(2, 11), silhouette_scores)
plt.title('Silhouette method')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette Score')
plt.show()
```
