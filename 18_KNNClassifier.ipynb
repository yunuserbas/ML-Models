```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
titanic = sns.load_dataset('titanic')
titanic
```

```python
titanic.info()
```

```python
titanic.head(2)
```

```python
#On selectionne les colonnes qu'on va factorizer

strings=titanic.select_dtypes(include='object')
strings.nunique()
```

```python
titanic['facto_sex']=titanic['sex'].factorize()[0]
titanic['facto_alive']=titanic['alive'].factorize()[0]
titanic['facto_embarked']=titanic['embarked'].factorize()[0]
titanic['facto_who']=titanic['who'].factorize()[0]
titanic['facto_embark_town']=titanic['embark_town'].factorize()[0]
titanic['facto_alive']=titanic['alive'].factorize()[0]
```

```python
titanic
```

```python
#on selectionne les colonnes numeriques
titanic=titanic.select_dtypes(include=np.number)
titanic
```

```python
#on supprime les nan
titanic=titanic.dropna()
titanic.info()
```

```python
#on regarde la correlation des colonnes
corr_matrix = titanic.corr()
```

```python
corr_matrix
```

```python
#on supprime les colonnes redondantes 
titanic=titanic.drop(columns=['facto_alive','facto_embarked'], axis=1)
titanic
```

```python
X=titanic.drop(columns='survived')
y=titanic['survived']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
model=KNeighborsClassifier()
model.fit(X_train,y_train)
```

```python
print('train_score:',model.score(X_train,y_train))
print('test_score:',model.score(X_test,y_test))
```

```python
titanic['predict']=model.predict(X)
```
