```python
import numpy as np
import pandas as pd
import seaborn as sns

from sklearn.neighbors import KNeighborsClassifier
titanic = sns.load_dataset('titanic')
```

```python
titanic
```

```python
titanic.info()
```

```python
titanic['who'].value_counts()
```

```python
titanic['facto_who']=titanic['who'].factorize()[0]
```

```python
titanic['facto_class']=titanic['class'].factorize()[0]
```

```python
titanic['facto_sex']=titanic['sex'].factorize()[0]
```

```python
titanic['facto_embark_town']=titanic['embark_town'].factorize()[0]
```

```python
titanic
```

```python
titanic.drop(columns=["deck"], inplace = True)
```

```python
titanic.info()
```

```python
titanic.dropna(subset=['age'],inplace=True)
```

```python
titanic.dropna(subset=['embarked'],inplace=True)
```

```python
titanic.info()
```

```python
titanic.select_dtypes(include="number")
```

```python
X = titanic.select_dtypes(include="number").drop(columns= 'survived')
y = titanic['survived']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,train_size= 0.75, random_state=42)
```

```python
modelknn = KNeighborsClassifier()
```

```python
modelknn.fit(X_train,y_train)
```

```python
print('train : ',modelknn.score(X_train,y_train))
print('test : ',modelknn.score(X_test,y_test))
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=0.95)
```

```python
data_pca=pca.fit_transform(X)
```
